```python
# 数据导入
import pandas as pd
data = pd.read_csv('file:///home/hadoop/jupyternotebook/user_meal_rating.csv')
data.head(10)
```

```python
# 查询缺失值
data.isnull().sum()
```

```python
# 查询重复值
data.duplicated(subset=['UserID','Rating','MealID']).sum()
```

```python
# 使用ReviewTime对字段升序排序
data=data.sort_values(by=['ReviewTime'])
data.head(10)
```

```python
# 删除重复值仅保留最新时间戳数据
data.drop_duplicates(inplace=True)
data.head(10)
```

```python
# 再次查重
data.duplicated().sum()
```

```python
# 保存数据
data.to_csv('user_meal.csv',index=False)
```
